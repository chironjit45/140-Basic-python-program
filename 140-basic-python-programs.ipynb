## Problem 4 : Write a program to swap two variables. 

```python
# input two variables
a = input("Enter the first variables: ")
b = input("Enter the second variables: ")

# Display the values original values
print(f"Original values: a={a}, b={b}")

# Swap the values using a temporary variable 
temp = a 
a = b 
b = temp

#Display the swapped values
print(f"Swapped values: a={a}, b={b}")
```

## Program 9: Write a program to solve a quadratic equation. 

The standard form of a quadratic equation is: 
                                          ax^2 + bx + c = 0
                                where a,b and c are real number and a!=0 
The solution of this quadratic equation is given by: 
                                    (-b +- (b^2 - 4*a*c)^1/2) / 2a 

```python
import math

# input coefficients 
a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))

# calcualte the discriminant. 
discriminant = b**2 - 4*a*c 

if discriminant > 0:
    root1 = (-b + math.sqrt(discriminant))/(2*a)
    root2 = (-b - math.sqrt(discriminant))/(2*a)
    print(f"Root1: {root1}")
    print(f"Root2: {root2}")

elif discriminant == 0: 
    #one real root(repeated)
    root = -b / (2*a)

else: 
    #complex roots
    real_part = -b/(2*a)
    imaginary_part = math.sqrt(abs(discriminant))/(2*a)
    print(f"Root1: {real_part} + {imaginary_part}i")
    print(f"Root1: {real_part} - {imaginary_part}i")
    
```

## Problem 13:  Write a program to check for leap years. 

```python
year = int(input("Enter a year: "))
# Divided by 100 means century year(Ending with 00)
# century year divided by 4000 is leap year. 

if (year % 100==0) and (year % 400==0):
    print(f"{year} is leap year.")

# Not divided by 100 means not a century year. 
# year divided by 4 is a leap year.
elif (year % 100!=0) and (year%4==0):
    print(f"{year} is leap year")
else:
    print(f"{year} is not a leap year.")
```
